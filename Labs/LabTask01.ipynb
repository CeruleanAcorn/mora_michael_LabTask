### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



```python
#1. Convert categorical values in the gender column to numeric codes.
print("1. Convert categorical values in the gender column to numeric codes.")

def sex_categories(x):
    if(x == 'male'):
        return 0
    elif(x == 'female'):
        return 1
    else: # If input is something else, account for it with NaN
        return np.nan

stringgender = df[['Sex']]
# 0 = 'male', otherwise 1.
numericgender = stringgender.map(sex_categories)
numericgender
```

```python
#2. Create a new column that categorizes age into groups (e.g., child, adult, senior).
print("2. Create a new column that categorizes age into groups (e.g., child, adult, senior).")

def age_categories(x):
    if(x >=65):
        return 'Senior'
    elif(x >= 19):
        return 'Adult'
    elif(x >= 0):
        return 'Child'
    else: # Some inputs are NaN; this will account for that.
        return np.nan

ages = df[['Age']]

# Define all columns, adding 'Age Category' to the mix.
new_col = ['PassengerId', 'Survived', 'Pclass', 'Name',	'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked', 'Age Category']

# By starting with_age_category as a copy of df, we can print out the whole datasheet with the new Age Category column, not just that column.
with_age_category = df.reindex(columns=new_col)

# agecategory = with_age_category[['Age Category']]

# Age Category will start as a copy of Age's values.
with_age_category['Age Category'] = with_age_category[['Age']]

#Then, send with_age_category only to the age_categories function.
with_age_category['Age Category'] = with_age_category['Age Category'].map(age_categories)

with_age_category
```
