### Part 3: Filtering ###

1. Show only the rows where the age is above 60.

2. Show rows where the fare is above 100 and the passenger was in 1st class.

```python
# 1. Show only the rows where the age is above 60.
print("1. Show only the rows where the age is above 60.") 

abovesixtydf = df[df['Age'] > 60]
abovesixtydf
```

```python
# 2. Show rows where the fare is above 100 and the passenger was in 1st class.
print("2. Show rows where the fare is above 100 and the passenger was in 1st class.")

# 1st class appears to be represented by "1" in Pclass
highfarefirstclassdf = df[(df['Fare'] > 100) & (df['Pclass'] == 1)]
highfarefirstclassdf
```
