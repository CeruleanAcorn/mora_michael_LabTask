### Monthly streams

```python
def createHistogram(columnName, filterData=None):
    tempDf = df
    if filterData:
        tempDf = df[df[columnName] < 100]
    fig, ax= plt.subplots(figsize=(9,4))
    median = df.streams_millions.median()
    mean = df.streams_millions.mean()
    sns.histplot(data=df[df.streams_millions < 100], x="streams_millions", bins=20, ax=ax, color="steelblue")
    ax.axvline(median, color="darkorange", label="median")
    ax.axvline(mean, color="darkred", label="mean")
    ax.legend(frameon=False)
    ax.set_xlabel('Monthly stream (mllions)')
    ax.set_ylabel('Song count')
    ax.set_title('Monthly Streams')
    plt.show()
```

Bonus point: Histogram for another column!

```python
fig, ax= plt.subplots(figsize=(9,4))
median = df.danceability.median()
mean = df.danceability.mean()
sns.histplot(data=df[df.danceability < 100], x="danceability", bins=20, ax=ax, color="purple")
ax.axvline(median, color="green", linestyle="--", label=f"median: {median:.2f}")
ax.axvline(mean, color="yellow", linestyle="--",  label=f"mean: {mean:.2f}")
ax.legend(frameon=False)
ax.set_xlabel('Danceability')
ax.set_ylabel('Song count')
ax.set_title('Danceability of Songs')
plt.show()
```

```python
genres = ["Pop", "Hip-Hop", "Rock", "Electronic", "R&B", "Indie"]

trend = df[df.genre.isin(genres)].groupby(["year","genre"])["streams_millions"].median().reset_index()

fig, axes = plt.subplots(2,3,figsize=(12,5), sharex=True, sharey=True) #Prevents overlapping of x-axis content, share\Y also possible
axes = axes.flatten()

for ax, genre in zip(axes, genres):
    subD = trend[trend.genre == genre]
    # peak = subD.loc[subD.streams_millions.idxmax()]# Find x where y is the maximum!
    # print(peak)
    # break
    ax.annotate(f"peak: {peak.streams_millions:.1f} M", xy=(peak.year, peak.streams_millions), xytext=(20, 20), textcoords="offset points", ha="center", fontsize=8, arrowprops=dict(arrowstyle="->", color="black", lw=2)) #ha=hori alignment
    ax.plot(subD.year, subD.streams_millions, marker="o", ms=5)
    ax.set_title(genre, fontweight="bold")
#ax.plot(pop_trend.index, pop_trend)
```

```python
labels = ["Sony Music", "Warner Music", "A24 Sounds", "Concord"]

trend = df[df.label.isin(labels)].groupby(["year","label"])["streams_millions"].median().reset_index()

fig, axes = plt.subplots(2,2,figsize=(12,5), sharex=True, sharey=True) #Prevents overlapping of x-axis content, share\Y also possible
fig.suptitle("Streams in Millions for Record Label Music Over The Years", fontweight="bold", color="red")
axes = axes.flatten()

for ax, label in zip(axes, labels):    
    subD = trend[trend.label == label]
    # peak = subD.loc[subD.streams_millions.idxmax()]# Find x where y is the maximum!
    # print(peak)
    # break
    ax.annotate(f"peak: {peak.streams_millions:.1f} M", xy=(peak.year, peak.streams_millions), xytext=(40, 0), textcoords="offset points", ha="center", fontsize=10, arrowprops=dict(arrowstyle="->", color="red", lw=2)) #ha=hori alignment
    if ax == axes[0]: # Sony Music has the highest peak of all graphs shown, so have it in a different colour!
        ax.plot(subD.year, subD.streams_millions, marker="X", ms=8, color="blue")
    else:
        ax.plot(subD.year, subD.streams_millions, marker="X", ms=8, color="green")
    ax.set_title(label, fontweight="bold")
```

**Let's try making a basic pie chart in Altair!**

```python
import altair as alt

# Adapted from: https://altair-viz.github.io/gallery/pie_chart.html
source = pd.DataFrame({
    "category": df['genre'],
    "value": df['streams_millions']
})

# Adding title for pie chart from: https://altair-viz.github.io/user_guide/customization.html
alt.Chart(source, title="Distribution of Music Streams Per Millions, Per Genre").mark_arc().encode(
    theta="value:Q",
    color="category:N",
    # Adapted from https://www.educative.io/answers/how-to-draw-a-pie-chart-in-altair
    tooltip=['category', 'value'] # Allows mouse hover for viewing indiviual row entres and their contributions to the pie chart
)
```
